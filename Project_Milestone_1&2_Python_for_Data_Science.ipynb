1. Load the COVID-19 dataset using pandas from the provided dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the COVID-19 dataset
url = "https://lms.educlaas.com/assets/courseware/v1/e206258dd748e86b8bb35fc039e90fa8/asset-v1:EduCLaSS+PYD+PDDS-PYD-0525-21Jul2025C+type@asset+block/owid-covid-data.csv"
covid = pd.read_csv(url)
```

3. Check for missing values in the dataset and decide on a strategy to handle them.

```python
# 3. Check for missing values
missing_values = covid.isnull().sum()
print("\nMissing values:\n", missing_values[missing_values > 0])
```

4. Remove the columns with more than 90% of missing values.

```python
# 4. Remove columns with >90% missing. Strategy: Drop columns with >90% missing, then fill remaining
threshold = 0.9 * len(covid)
covid = covid.loc[:, covid.isnull().sum() <= threshold]
covid = covid.fillna(method='ffill').fillna(method='bfill')  # forward and backward fill
covid.info()
```

5. Convert the 'date' column to the datetime data type.

```python
# 5. Convert 'date' column to datetime
covid['date'] = pd.to_datetime(covid['date'], errors='coerce')
```

# Activity 2

1. Impute missing values in the dataset columns

```python
# once we confirmed the missing were genuine 0 for time_series
time_series = [ 'total_cases', 'total_deaths', 'total_cases_per_million',  'total_tests', 'people_vaccinated', 'new_cases', 'new_deaths']
for col in time_series:
   covid[col] = covid[col].fillna(0)
```

```python
non_time_series = ['positive_rate','male_smokers','female_smokers','people_vaccinated', 'hospital_beds_per_thousand','population_density', 'life_expectancy']
for col in non_time_series:
    covid[col] = covid.groupby('location')[col].transform(lambda x: x.fillna(x.mean()))
    covid[col] = covid[col].fillna(0)
```

```python
# replace object column missing values with string  "Unknown"
covid['continent'] = covid['continent'].fillna('Unknown')
```

```python
# 5. Impute missing values using forward and backward fill
# covid = covid.fillna(method='ffill').fillna(method='bfill')
```

2. Remove duplicate rows from the DataFrame.

```python
# 6. Remove duplicate rows
covid = covid.drop_duplicates()
```

3. Create new features if needed (e.g., extract year and month from the 'date' column).

```python
# 7. Create new features from date
covid['year'] = covid['date'].dt.year
covid['month'] = covid['date'].dt.month
```

4. Explore unique countries in the dataset and count the total number of countries.

```python
# 8. Explore unique countries and count
unique_countries = covid['location'].dropna().unique()
print("\nUnique countries:\n", unique_countries)
print("\nTotal number of countries:", len(unique_countries))
```

# Activity 3

1. Visualize the WHO Regions (Continent column in dataset) with total COVID-19 cases and total deaths by using bar plots. .

```python
# Drop rows with missing continent values
covid = covid.dropna(subset=['continent'])

# Fill missing numeric values with 0
covid['total_cases'] = covid['total_cases'].fillna(0)
covid['total_deaths'] = covid['total_deaths'].fillna(0)

# Group by continent and take the maximum values
region_summary = covid.groupby('continent')[['total_cases', 'total_deaths']].max().reset_index()

# Plot total cases
plt.figure(figsize=(10, 5))
sns.barplot(data=region_summary, x='continent', y='total_cases')
plt.title("Total COVID-19 Cases by WHO Region (Continent)")
plt.xlabel("Continent")
plt.ylabel("Total Cases")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot total deaths
plt.figure(figsize=(10, 5))
sns.barplot(data=region_summary, x='continent', y='total_deaths')
plt.title("Total COVID-19 Deaths by WHO Region (Continent)")
plt.xlabel("Continent")
plt.ylabel("Total Deaths")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

2. Explore the world wide monthly trend of  COVID-19 cases using a line plot. .

```python
# Group by year and month to sum new cases
monthly_trend = covid.groupby(['year', 'month'])['new_cases'].sum().reset_index()

# Create a 'year_month' column for plotting
monthly_trend['year_month'] = pd.to_datetime(monthly_trend[['year', 'month']].assign(day=1))

# Plotting the line chart
plt.figure(figsize=(14, 6))
sns.lineplot(data=monthly_trend, x='year_month', y='new_cases')
plt.title("Worldwide Monthly Trend of New COVID-19 Cases")
plt.xlabel("Month")
plt.ylabel("New Cases")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

3. Investigate the correlation between total cases and total deaths using a heatmap.

```python
# Compute the correlation matrix
correlation_matrix = covid[['total_cases', 'total_deaths']].corr()

# Plot the heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Between Total Cases and Total Deaths")
plt.tight_layout()
plt.show()
```

4. Analyze how total cases have evolved over time for a specific location (e.g., India).

```python
# Filter data for India and sort by date
india_data = covid[covid['location'] == 'India'].sort_values('date')

# Drop missing values in 'total_cases'
india_data = india_data.dropna(subset=['total_cases'])

# Plot the line chart
plt.figure(figsize=(12, 6))
sns.lineplot(data=india_data, x='date', y='total_cases')
plt.title("Total COVID-19 Cases Over Time in India")
plt.xlabel("Date")
plt.ylabel("Total Cases")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Task 2: Visualize Passenger Survival

*  Create a bar plot using seaborn to visualize the count of survivors and non-survivors.

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
url = "https://raw.githubusercontent.com/angseesiang/lithan/main/titanic.csv"
titanic = pd.read_csv(url, index_col=0)

# Set plot style
sns.set(style="whitegrid")

# Create the bar plot
sns.countplot(data=titanic, x='Survived')

# Add labels and title
plt.xlabel('Survival (0 = No, 1 = Yes)')
plt.ylabel('Passenger Count')
plt.title('Count of Survivors and Non-Survivors')

# Show the plot
plt.show()
```

Task 3: Analyze Passenger Age Distribution

*  Create a histogram using matplotlib to show the distribution of passenger ages.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
url = "https://raw.githubusercontent.com/angseesiang/lithan/main/titanic.csv"
titanic = pd.read_csv(url, index_col=0)
# Plot histogram of Age
plt.figure(figsize=(10, 6))
plt.hist(titanic['Age'].dropna(), bins=30, edgecolor='black', color='skyblue')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.title('Distribution of Passenger Ages')

# Show the plot
plt.show()
```

Task 4: Explore Class-wise Distribution

*  Use seaborn to create a count plot to display the number of passengers in each class.

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
url = "https://raw.githubusercontent.com/angseesiang/lithan/main/titanic.csv"
titanic = pd.read_csv(url, index_col=0)

# Set seaborn style
sns.set(style="whitegrid")

# Create count plot for passenger class
plt.figure(figsize=(8, 5))
sns.countplot(data=titanic, x='Pclass', palette='pastel')

# Add labels and title
plt.xlabel('Passenger Class (Pclass)')
plt.ylabel('Number of Passengers')
plt.title('Number of Passengers by Class')

# Show the plot
plt.show()
```

Task 6: Correlation Between Variables

*  Generate a heatmap using seaborn to visualize the correlation matrix of the numerical variables in the dataset.

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Titanic dataset
url = "https://raw.githubusercontent.com/angseesiang/lithan/main/titanic.csv"
titanic = pd.read_csv(url, index_col=0)

# Compute the correlation matrix (numerical columns only)
corr_matrix = titanic.corr(numeric_only=True)

# Set up the matplotlib figure
plt.figure(figsize=(10, 8))

# Create the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

# Add title
plt.title('Correlation Matrix of Numerical Variables')

# Show the plot
plt.show()
```

Task 7: Visualize Passenger Class and Age

*  Create a scatter plot using matplotlib to show the relationship between passenger age and class.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the Titanic dataset
url = "https://raw.githubusercontent.com/angseesiang/lithan/main/titanic.csv"
titanic = pd.read_csv(url, index_col=0)

# Drop rows with missing Age or Pclass
data = titanic[['Age', 'Pclass']].dropna()

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(data['Age'], data['Pclass'], alpha=0.6, color='teal', edgecolors='k')

# Label axes and add title
plt.xlabel('Age')
plt.ylabel('Passenger Class (Pclass)')
plt.title('Scatter Plot of Passenger Age vs Class')
plt.grid(True)

# Invert y-axis so 1st class appears on top (optional, for intuitive class ranking)
plt.gca().invert_yaxis()

# Show the plot
plt.show()
```
